### Experimento 1

Para una corriente de control de 0.52 $A$ se obtuvo la siguiente gráfica:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

## Creamos dos arreglos con los datos que obtuvimos en la práctica
corriente = [0.000, 0.106, 0.204, 0.301, 0.404, 0.502, 0.607, 0.707, 0.800, 0.896, 0.998]  ## Corriente medida en Amperes (A)
voltaje_hall = [0.1, 0.9, 1.8, 2.7, 3.9, 4.9, 6.3, 7.0, 7.9, 8.7, 9.8]  # Voltaje de Hall medido en mV

## Realizamos la regresión lineal de los datos anteriormente ingresado
slope, intercept, r_value, p_value, std_err = linregress(corriente, voltaje_hall)

## Después calculamos la línea de regresión 
regression_line = slope * np.array(corriente) + intercept

## Finalmente graficamos los datos y la línea de regresión correspondiente a la gráfica
plt.figure(figsize=(8, 6))
plt.plot(corriente, voltaje_hall, 'bo', label='Datos experimentales')
plt.plot(corriente, regression_line, 'r-', label=f'Regresión lineal: y={slope:.2f}x + {intercept:.2f}')
plt.title('Voltaje de Hall vs Corriente')
plt.xlabel('Corriente (A)')
plt.ylabel('Voltaje de Hall (mV)')
plt.legend()
plt.grid(True)
plt.show()

## Para terminar calculamos e imprimimos el valor correspondiente a R^2
r_squared = r_value**2
print(f'Valor de R^2: {r_squared:.4f}')
```

Nota: en escencia todos los códigos subsecuentes a este experimento siguen la misma estructura, pero igual los comentaré por si se necesita

### Ejercicio 3

Unicamente graficando el conjunto de datos que obtuvimos (grados vs voltaje de Hall) siguen la siguiente distribución

```python
import numpy as np
import matplotlib.pyplot as plt

## Creamos los arreglos donde almacenamos las medidas obtenidas en el experimento
angulo = np.array([0.0, 15.0, 30.0, 45.0, 60.0, 75.0, 90.0, 105.0, 120.0, 135.0, 150.0, 165.0, 180.0, 195.0, 210.0, 225.0, 240.0, 255.0, 270.0, 285.0, 300.0, 315.0, 330.0, 345.0, 360.0])
voltaje_hall = np.array([0.1, 3.5, 7.7, 11.1, 15.6, 18.1, 20.1, 18.1, 16.0, 12.1, 8.5, 4.3, 0.2, -5.1, -9.6, -11.5, -17.4, -18.5, -21.5, -19.9, -17.2, -15.1, -10.0, -5.4, 0.2])

# Graficar solo los puntos
plt.figure(figsize=(12, 8))
plt.plot(angulo, voltaje_hall, 'bo', marker='o', linestyle='None', color='blue', label='Datos')

plt.xlabel('Ángulo (°)')
plt.ylabel('Voltaje Hall (mV)')
plt.title('Ángulo vs Voltaje Hall')
plt.legend()
plt.grid(True)
plt.show()
```

En principio es visible que los datos efectivamente siguen (o se aproximan) a una función senoidal, para comprobar esto al mismo código se le hará una modificación para que a la par del conjunto de datos experimentalmente obtenidos tambien se grafique la función $\sin(x)$ en un intervalo $x \in [0,2\pi] $ (o su equivalente en grados de $0^°$ a $360^°$) para comparar ambas distribuciones

```python
import numpy as np
import matplotlib.pyplot as plt

## Nuevamente creamos nuestros arreglos que contengan los datos experimentalmente obtenidos
angulo = np.array([0.0, 15.0, 30.0, 45.0, 60.0, 75.0, 90.0, 105.0, 120.0, 135.0, 150.0, 165.0, 180.0, 195.0, 210.0, 225.0, 240.0, 255.0, 270.0, 285.0, 300.0, 315.0, 330.0, 345.0, 360.0])
voltaje_hall = np.array([0.1, 3.5, 7.7, 11.1, 15.6, 18.1, 20.1, 18.1, 16.0, 12.1, 8.5, 4.3, 0.2, -5.1, -9.6, -11.5, -17.4, -18.5, -21.5, -19.9, -17.2, -15.1, -10.0, -5.4, 0.2])

## Con esto unicamente creamos el gràfico de los puntos de la distribuciòn de los datos de arriba
plt.figure(figsize=(12, 8))
plt.plot(angulo, voltaje_hall, 'bo', marker='o', linestyle='None', color='blue', label='Datos')

## Ahora creamos la gràfica de la funciòn seno
## Para ello entonces generamos un rango de ángulos de 0 a 360 grados o 0 a 2pi si trabajaramos en radianes
angulo_fino = np.linspace(0, 360, 360)
angulo_fino_rad = np.radians(angulo_fino)

## Con eso entonces ya podemos definir nuestra funciòn seno
seno = np.sin(angulo_fino_rad) * (np.max(voltaje_hall) - np.min(voltaje_hall)) / 2 + np.mean(voltaje_hall)

plt.plot(angulo_fino, seno, 'r--', label='Función Seno')

plt.xlabel('Ángulo (°)')
plt.ylabel('Voltaje Hall (mV)')
plt.title('Ángulo vs Voltaje Hall comparado contra una Función Senoidal')
plt.legend()
plt.grid(True)
plt.show()
```

Y efectivamente nuestros datos se ajustan a la función seno por lo que la teoría de este experimento quedó demostrada, sin embargo es cierto que algunos puntos no estan del todo dentro de la función seno esto se debe a que la nuez graduadora que se uso durante el experimento para medir los angulos estaba en un mal estado sobre todo entre los grados $200^°$ y $300^°$ que es justamente donde tenemos la mayor variación respecto al ajuste senoidal.
