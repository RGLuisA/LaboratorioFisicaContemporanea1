```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

## Ingresamos los datos de la tabla experimental
potencia_sin_pdms = np.array([18.8, 20.4, 41.2, 45.7, 14.2])
potencia_con_pdms = np.array([12.16, 4.34, 14.2, 7.86, 2.62])

## Calculamos las perdidas con el valor de la ecuacion teorica
perdidas = 10 * np.log10(potencia_sin_pdms / potencia_con_pdms)

## Ingresamos las longitudes estimadas de PDMS en cm para cada medición
longitudes_pdms = np.array([0.5, 1.0, 1.5, 2.0, 2.5])

## Hacemos el ajuste lineal correspondiente a los datos experimentales
slope, intercept, r_value, p_value, std_err = stats.linregress(longitudes_pdms, perdidas)

## Creamos la gráfica de pérdidas en función de la longitud del PDMS
plt.figure(figsize=(8, 6))
plt.scatter(longitudes_pdms, perdidas, color='b', label="Pérdidas (dB)")  
plt.plot(longitudes_pdms, slope*longitudes_pdms + intercept, 'r-', label=f"Ajuste Lineal: $y = {slope:.2f}x + {intercept:.2f}$")
plt.xlabel("Longitud del PDMS (cm)")
plt.ylabel("Pérdidas (dB)")
plt.title("Gráfica de longitud del PDMS vs Pérdidas")
plt.grid(True)
plt.legend()
plt.tight_layout()

## Imprimimos los resultados importantes
print(f"Ecuación de la recta: y = {slope:.2f}x + {intercept:.2f}")
print(f"Coeficiente de determinación (R^2): {r_value**2:.4f}")
```

```python
## Imprimimos las perdidas individuales
for i, loss in enumerate(perdidas, start=1):
    print(f"Pérdida {i}: {loss:.2f} dB")
```
